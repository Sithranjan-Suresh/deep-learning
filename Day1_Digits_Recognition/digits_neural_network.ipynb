```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train.shape
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train_flattened = X_train.reshape(len(X_train), 28*28)
```

```python
X_train_flattened.shape
```

```python
X_test_flattened = X_test.reshape(len(X_test), 28*28)
X_test_flattened.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape= (784,), activation='sigmoid')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy',metrics=['accuracy'] )
model.fit(X_train_flattened, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattened, y_test)
```

```python
plt.matshow(X_test[0])
```

```python
y_predicted = model.predict(X_test_flattened)
```

```python
y_predicted[0]
```

```python
np.argmax(y_predicted[0])
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```python
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_labels)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

Now, let's add a hidden layer


```python
model_2 = keras.Sequential([
    keras.layers.Dense(100, input_shape= (784,), activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model_2.compile(optimizer='adam', loss='sparse_categorical_crossentropy',metrics=['accuracy'] )
model_2.fit(X_train_flattened, y_train, epochs=5)
```

```python
model_2.evaluate(X_test_flattened, y_test)
```

```python
y_predicted = model_2.predict(X_test_flattened)
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```python
cm2 = tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_labels)
cm2
```

```python

plt.figure(figsize = (10,7))
sn.heatmap(cm2, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
