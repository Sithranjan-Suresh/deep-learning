Sigmoid function:

```python
import math

def sigmoid(x):
  if x >= 0:
    return 1 / (1 + math.exp(-x))
  else:
    # Use the mathematically equivalent form e^x / (1 + e^x)
    # This avoids computing math.exp(large_positive_number)
    exp_x = math.exp(x)
    return exp_x / (1 + exp_x)
```

```python
sigmoid(1000)
```

```python
sigmoid(-1000)
```

Tanh function:

```python
def tanh(x):
  return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
tanh(-56)
```

```python
tanh(100)
```

ReLU function:

```python
def relu(x):
    return max(0,x)
```

```python
relu(100)
```

```python
relu(-100)
```

Leaky ReLU function:

```python
def leaky_relu(x):
  return max(0.1*x, x)
```

```python
leaky_relu(-100)
```

```python
leaky_relu(8)
```
