```python
import numpy as np

y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.3,0.7,1,0,0.5])
```

```python
def mae(y_true, y_predicted):
  total_error = 0
  for yt, yp in zip(y_true, y_predicted):
    total_error += abs(yt-yp)
  print("Total error: ", total_error)
  mean = total_error/len(y_true)
  print("MAE: ", mean)
```

```python
mae(y_true, y_predicted)
```

Same function but using numpy

```python
np.mean(np.abs(y_predicted - y_true))
```

```python
epsilon = 1e-15
```

```python
y_predicted_new = [epsilon if i == 0 else (1 - epsilon if i == 1 else i) for i in y_predicted]
y_predicted_new
```

```python
y_predicted_new = np.array(y_predicted_new)
np.log(y_predicted_new)
```

```python
-np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def log_loss(y_true, y_predicted):
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
log_loss(y_true, y_predicted)
```

Exercise: Implement mean squared error
1) Without numpy

2) With numpy

```python
#Without numpy

def mse_without(y_true, y_predicted):
  error = 0
  for yt,yp in zip(y_true, y_predicted):
    error += (yt-yp)**2
  print("Total error: ", error)
  print("MSE: ", (error / len(y_true)))
```

```python
mse_without(y_true, y_predicted)
```

```python
#With numpy
def mse_with(y_true, y_predicted):
  error = 0
  error = np.sum(np.square(y_true-y_predicted))
  print("Total error: ", error)
  print("MSE: ", (error / len(y_true)))
```

```python
mse_with(y_true, y_predicted)
```
